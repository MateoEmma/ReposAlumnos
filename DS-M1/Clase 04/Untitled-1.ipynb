```python
import pandas as pd
```

```python
hospitales = pd.read_csv('hospitales2.csv')
```

```python
hospitales.head()
```

```python
hospitales1 = hospitales.copy()
```

```python
hospitales1 =hospitales1[['WKT', 'NOMBRE', 'DOM_NORMA']]
hospitales1.head()
```

```python
from matplotlib.pyplot import axis



hospitales1['WKT']= hospitales1['WKT'].str.replace('(','')
hospitales1['WKT']= hospitales1['WKT'].str.replace(')','')
hospitales1['WKT']= hospitales1['WKT'].str.replace('POINT', '')
hospitales1['LONGITUD'] =hospitales1.WKT.str.split(pat= ' ', expand=True)[1]
hospitales1['LATITUD']= hospitales1.WKT.str.split(pat= ' ', expand=True)[2]
del(hospitales1['WKT'])
hospitales1= hospitales1.reindex(columns=['LATITUD','LONGITUD','NOMBRE','DOM_NORMA'])
hospitales1.head()
```

```python
hospitales1.columns=['LATITUD','LONGITUD','LABEL','NAME']
hospitales1
```

```python
hospitales1.to_csv('Hospital2resuelto.csv')

```
